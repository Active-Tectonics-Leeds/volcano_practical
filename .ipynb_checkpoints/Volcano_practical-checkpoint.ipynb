## 0. Introduction

A Mogi model is a point-source model used to explain surface deformation above an expanding magma chamber.

We make the following assumptions:
- the source in in an isotropic elastic half space, which is defined by the Poisson's ratio $v$, and the Shear modulus $\mu$.
- the radius of the point source is lower than it depth (can't break the surface).
- the magma itself is incompressible.

We can express a mogi source as:

\begin{equation}
\begin{pmatrix} ux \\ uy \\ uz \end{pmatrix} = \alpha^3 \delta P \frac{(1-v)}{\mu} \begin{pmatrix} x / R^3 \\ y / R^3 \\ z / R^3 \end{pmatrix} \qquad \qquad (1)
\end{equation}

By taking the following approximation:
\begin{equation}
\delta V \approx \frac{\delta P}{\mu} \pi \alpha^3
\end{equation}

We can rewrite Equation 1 as:

\begin{equation}
\begin{pmatrix} ux \\ uy \\ uz \end{pmatrix} = \delta V \frac{(1-v)}{\pi} \begin{pmatrix} x / R^3 \\ y / R^3 \\ z / R^3 \end{pmatrix} \qquad \qquad (2)
\end{equation}

## 1. Synthetic Mogi Source

Let's start by generating a simple models for an imflating magma chamber - a mogi source.

```python
# setup coord limits in km
xmin, xmax, xint = -20, 20, 0.1
ymin, ymax, yint = -20, 20, 0.1

# setup model params
source_depth = 3 # depth to point (km)
dV = 1e6 # change in volume (m^3)
nu = 0.25 # poisson's ratio for medium

# make grids of x and y coordinates
x = np.arange(xmin, xmax+xint, xint)
y = np.arange(ymin, ymax+yint, yint)
xx, yy = np.meshgrid(x, y)

# generate displacements using Mogi source (note the unit conversions to metres)
ux, uy, uz = lib.mogi_source(xx*1000, yy*1000, d=source_depth*1000, dV=dV, nu=nu)

# plot the displacements
fig, axs = plt.subplots(1,3,figsize=(20,10))
fig.tight_layout()

im = axs[0].imshow(ux*1000, extent=[xmin, xmax, ymin, ymax])
divider = make_axes_locatable(axs[0])
cax = divider.append_axes("bottom", size="7%", pad="6%")
plt.colorbar(im, cax=cax, orientation="horizontal", label='Displacement (mm)')
axs[0].set_title('Displacement in x')
axs[0].xaxis.tick_top()

im = axs[1].imshow(uy*1000, extent=[xmin, xmax, ymin, ymax])
divider = make_axes_locatable(axs[1])
cax = divider.append_axes("bottom", size="7%", pad="6%")
plt.colorbar(im, cax=cax, orientation="horizontal", label='Displacement (mm)')
axs[1].set_title('Displacement in y')
axs[1].set_yticklabels([])
axs[1].set_yticks([])
axs[1].xaxis.tick_top()

im = axs[2].imshow(uz*1000, extent=[xmin, xmax, ymin, ymax])
divider = make_axes_locatable(axs[2])
cax = divider.append_axes("bottom", size="7%", pad="6%")
plt.colorbar(im, cax=cax, orientation="horizontal", label='Displacement (mm)')
axs[2].set_title('Displacement in z')
axs[2].yaxis.tick_right()
axs[2].xaxis.tick_top()
```

Let's also plot these are a 3D surface displacement map.

## 2. Impact of Satellite Line-of-Sight

The Sentinel-1 SAR Satellites use side-looking instruments. Also, measurements from InSAR are one-dimensional in the satellite's line-of-sight. These characteristics means that the displacements measured by InSAR are different to those generated by our Mogi model.

Let's run the model again but include the effect of LOS, starting with a single value for both the satellite heading and the incidence angle.

```python
# params
heading = 147 # degrees
incidence = 45 # degrees

# create grids of these angles
heading_grid = np.ones(xx.shape) * heading
incidence_grid = np.ones(xx.shape) * incidence

# generate conversions for each components to LOS
x2los = np.sin(np.deg2rad(heading_grid)) * np.sin(np.deg2rad(incidence_grid))
y2los = np.cos(np.deg2rad(heading_grid)) * np.sin(np.deg2rad(incidence_grid))
z2los = -np.cos(np.deg2rad(incidence_grid))

# apply projection
ux_los = ux * x2los
uy_los = uy * y2los
uz_los = uz * z2los

# calculate total displacement in los 
ulos = ux_los + uy_los + uz_los

# plot our new displacements
fig, axs = plt.subplots(2,2,figsize=(10,10))
fig.tight_layout(w_pad=5)

im = axs[0,0].imshow(ux_los*1000, extent=[xmin, xmax, ymin, ymax])
divider = make_axes_locatable(axs[0,0])
cax = divider.append_axes("right", size="5%", pad="5%")
plt.colorbar(im, cax=cax, label='Displacement (mm)')
axs[0,0].set_title('Displacement in x')

im = axs[0,1].imshow(uy_los*1000, extent=[xmin, xmax, ymin, ymax])
divider = make_axes_locatable(axs[0,1])
cax = divider.append_axes("right", size="5%", pad="5%")
plt.colorbar(im, cax=cax, label='Displacement (mm)')
axs[0,1].set_title('Displacement in y')

im = axs[1,0].imshow(uz_los*1000, extent=[xmin, xmax, ymin, ymax])
divider = make_axes_locatable(axs[1,0])
cax = divider.append_axes("right", size="5%", pad="5%")
plt.colorbar(im, cax=cax, label='Displacement (mm)')
axs[1,0].set_title('Displacement in z')

im = axs[1,1].imshow(ulos*1000, extent=[xmin, xmax, ymin, ymax])
divider = make_axes_locatable(axs[1,1])
cax = divider.append_axes("right", size="5%", pad="5%")
plt.colorbar(im, cax=cax, label='Displacement (mm)')
axs[1,1].set_title('Displacement in LOS')

plt.show()
```

The displacements have shifted from (0,0), although this is a little hard to see in map-view like this.
Let's try taking some simple profiles through the displacement maps so we can compare with and without the LOS projection.

```python
# total displacement without los
u_total = np.sqrt(ux**2 + uy**2 + uz**2)

# get indexes of x=0 and y=0
x0_ind = np.argmin(x)
y0_ind = np.argmin(y)

# profile across x=0 using simple indexing
prof_y_without_los = u_total[:,x0_ind]
prof_y_with_los = ulos[:,x0_ind]

# and now across y=0
prof_x_without_los = u_total[y0_ind,:]
prof_x_with_los = ulos[y0_ind,:]

# plot profiles
fig, axs = plt.subplots(2,1,figsize=(10,10))
plt.tight_layout(h_pad=5)

axs[0].plot(x, prof_x_without_los*1000, color='red', label='Without LOS projection')
axs[0].plot(x, prof_x_with_los*1000, color='blue', label='With LOS projection')
axs[0].set_title('Profiles in x at y=0')
axs[0].legend()
axs[0].set_xlabel('Distance (km)')
axs[0].set_ylabel('Displacement (mm)')

axs[1].plot(x, prof_y_without_los*1000, color='red', label='Without LOS projection')
axs[1].plot(x, prof_y_with_los*1000, color='blue', label='With LOS projection')
axs[1].set_title('Profiles in y at x=0')
axs[1].legend()
axs[1].set_xlabel('Distance (km)')
axs[1].set_ylabel('Displacement (mm)')

plt.show()
```
